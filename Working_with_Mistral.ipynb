```python
from mistralai import Mistral, UserMessage

def mistral(user_message, model="mistral-large-latest"):
    client = Mistral(api_key=api_key)

    messages = [
        UserMessage(content=user_message),
    ]

    chat_response = client.chat.complete(
        model=model,
        messages=messages,
    )

    return chat_response.choices[0].message.content
```

```python
print(mistral("Hello, what can you do?"))
```

```python
prompt = """
You are a bank customer service bot.
Classify the inquiry into one of:
card arrival
change pin
exchange rate
country support
cancel transfer
charge dispute
customer service

Inquiry: I am inquiring about the availability of your cards in the EU.
Category:
"""

print(mistral(prompt))
```
